```python
import pandas as pd
```

```python
from jupyter_dash import JupyterDash
```

```python
from dash import dcc, html, Input, Output
```

```python
import plotly.express as px
```

```python
df = pd.read_csv("/content/Lung Cancer (1).csv")
```

```python
df['diagnosis_date'] = pd.to_datetime(df['diagnosis_date'])
```

```python
df.head()
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    html.H1("Lung Cancer Dashboard", style={'textAlign': 'center'}),

    html.Label("Select Country"),
    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': c, 'value': c} for c in df['country'].unique()],
        value=df['country'].unique()[0]
    ),

    dcc.Graph(id='age-distribution'),
    dcc.Graph(id='bmi-cholesterol'),
    dcc.Graph(id='cancer-stage')
])
```

```python
@app.callback(
    Output('age-distribution', 'figure'),
    Output('bmi-cholesterol', 'figure'),
    Output('cancer-stage', 'figure'),
    Input('country-dropdown', 'value')
)
def update_graphs(selected_country):
    filtered_df = df[df['country'] == selected_country]

    fig1 = px.histogram(
        filtered_df, x='age', color='gender',
        title='Age Distribution by Gender'
    )

    fig2 = px.scatter(
        filtered_df, x='bmi', y='cholesterol_level',
        color='smoking_status',
        title='BMI vs Cholesterol Level by Smoking Status'
    )

    fig3 = px.pie(
        filtered_df, names='cancer_stage',
        title='Cancer Stage Distribution'
    )

    return fig1, fig2, fig3
```

```python
app.run_server(mode='external')
```
